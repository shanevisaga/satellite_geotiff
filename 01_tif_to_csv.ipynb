```python
import pandas as pd
import xarray as xr
```

```python

#if you have multiple sites/ countries you can just add them in this list
#for loc in ['Pampanga','PH','Metro_Manila','Cebu']: 
for loc in ['PH']:
    for year in range(2019,2020): #edit this to range(year_start,year_last+1); this is currently set to only look for year 2019 files
        fp = f'{loc}_0317_0515_{year}' #NO2 filenames; the for loop can go through all your input files with varying countrynames 
        #opening the geotiff files and preparing csv files version
        dataarray = xr.open_rasterio(f'{fp}.tif')
        df = dataarray[0].to_pandas()
        df_list = pd.DataFrame(df).stack().rename_axis(['y', 'x']).reset_index(name='val')
        df_list.to_csv(f'{fp}.csv')

        #this part will just show a quickview of your geotiff files 
        #can be removed if sure that geotiff files are working fine
        import rasterio
        from rasterio.plot import show

        img = rasterio.open(f'{fp}.tif')
        show(img)
```
